```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime

# Generate dummy data with 1456 samples and 400 features
num_samples = 1466
num_features = 400
X_train = np.random.randn(num_samples, num_features)  # 1456 samples, 400 features
y_train = np.random.randn(num_samples, 1)             # 1456 target values

# Define your model architecture
def create_model():
    model = Sequential([
        Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(5, activation='softmax')
    ])
    model.compile(optimizer='Nadam', loss='mse', metrics=['accuracy'])
    return model

# Dictionary to store Mean Squared Errors (MSEs) for each batch size
mses = {}

# Define batch sizes
num_batch_sizes = [8, 12, 16, 32]

# Iterate over each batch size
for batch_size in num_batch_sizes:
    # Initialize model for each batch size
    model = create_model()
    
    # Callback for saving the best model
    checkpointer = ModelCheckpoint(filepath=f'saved_models/audio_classification_{batch_size}.hdf5', 
                                   verbose=1, save_best_only=True)
    start = datetime.now()

    # Fit the model for the current batch size
    model.fit(X_train, y_train, batch_size=batch_size, epochs=100, 
              validation_split=0.2, callbacks=[checkpointer], verbose=0)
    
    duration = datetime.now() - start
    print(f"Training completed for batch size {batch_size} in time: {duration}")
    
    # Load the best model for the current batch size
    model.load_weights(f'saved_models/audio_classification_{batch_size}.hdf5')
    
    # Generate test data (dummy data for demonstration)
    X_test = np.random.randn(1466, num_features)  # 200 samples for testing
    y_test = np.random.randn(1466, 1)            # 200 target values for testing
    
    # Predict on test data
    y_pred = model.predict(X_test, batch_size=batch_size)
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean(np.square(y_pred - y_test))
    
    # Store the MSE for the current batch size
    mses[batch_size] = mse

# Print MSEs
for batch_size, mse in mses.items():
    print(f"Batch size: {batch_size}, MSE: {mse}")
```
